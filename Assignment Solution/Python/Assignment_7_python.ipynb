Q.2.

Raw strings are used because of the backslash problem. The regex matching happens in Python using the re module. We write the regular expression as a string which is compiled by the module's compiler and then matched against a string.

Suppose we have a string with a backslash that we want to match: \sbc

The re module uses "\" for escaping a "\". So after compilation, the resulting expression should be "\\sbc".

But python also uses "\" for escaping its "\". So for each of the "\"s we have to add another "\". So the string pattern to be used will be "\\\\sbc".

s = "\\\\sbc" -> len(s) = 5

"\\\\sbc" gets converted to "\\sbc" after compilation.

So while checking the engine sees the first "\" as escaping the second "\". As a result we match with \sbc.

Thus, for matching one "\" we have to use 4 "\". This affects the readability.

Instead, we can use raw strings. It treats "\" just as a normal character, not an escaping sequence character.

"\\sbc" is equivalent to r"\sbc"

So to match \sbc, we can use the pattern r"\\sbc". Now, for Python, this string has a length 4.

```python
"\\sbc" == r"\sbc"
```

Q.5.

r'(\d\d\d)-(\d\d\d-\d\d\d\d)'

This is a raw string literal. So \d represents a decimal digit. The first group has three digits. In between the 2 groups, we have a hyphen. The second group has 3 digits followed by a hyphen followed by 4 digits. 
So if the pattern is abc-def-wxyz where a,b,c,d,e,f,w,x,y,z are decimal digits:

The group zero is the entire string that matched the regex i.e. "abc-def-wxyz"
Group 1 will be the first 3 digits i.e. "abc"
Group 2 will be the next 3 digits followed by the hyphen followed by the 4 digits i.e. "def-wxyz"

```python
m = re.search(r'(\d\d\d)-(\d\d\d-\d\d\d\d)', '999-200-2000')
m
```

```python
m.group(0)
```

```python
m.group(1)
```

```python
m.group(2)
```

```python
re.match(".*?","abcde").group()
```

```python
# Q.20.


regex = r"(\d{1,3},)?(?(1)(\d{3},)*|)(?(1)(\d\d\d$)|(\d\d?\d?$))"
```

```python
l = ["1", "10", "100", "1,000", "10,000", "100,000", "1,000,000", "10,000,000", "100,000,000", "1,000,000,000"]
l2 = [",", ",1", "1,", "10,", "1,0", "1,00", "10,0", "1000", "10,00", "100,0", "10,00,000", "10,000,000,"]
for x in l:
    assert re.match(regex, x) != None
print("No error")
```

```python
for x in l2:
    assert re.match(regex, x) == None
print("No error")
```
